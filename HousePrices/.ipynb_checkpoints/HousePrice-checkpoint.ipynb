```python
missTotal = (data.isnull().sum()).sort_values(ascending=False)
Counts = (data.isnull().count()).sort_values(ascending=False)
Percent = (missTotal / Counts)
```

```python
missData = pd.concat([missTotal,Percent],keys=['Total Missing Values','Percentage'],axis = 1)
```

```python
missData.head(20)
```

```python
data = data.drop((missData[missData['Percentage']>0.005].index),axis=1)
```

```python
data = data.drop((data.loc[data['Electrical'].isnull()].index),axis=0)
```

```python
(data.isnull().sum()).sort_values(ascending=False)
```

```python
stdscaler = StandardScaler()
stdsale = stdscaler.fit_transform(data['SalePrice'][:,np.newaxis])
```

```python
sns.distplot(stdsale) ##shows long tail on the right -- outliers.
```

```python
data.plot.scatter(x='SalePrice',y='GrLivArea') # Delete the outlier with high GrLivArea but low SalePrice.
```

```python
data= data.drop((data['GrLivArea'].sort_values(ascending=False)[:2]).index)
```

```python
data.plot.scatter(x='GrLivArea',y='SalePrice') #Now they are gone!
```

```python
sns.distplot(np.log(data['SalePrice']),fit=norm) # Lognorm distribution -- Typical distribution for data with only positive numbers.
```

```python
stats.probplot(np.log(data['SalePrice']),plot=plt)
```

```python
data['SalePrice'] = np.log1p(data['SalePrice'])
```

```python
#Find out all skewed numerical distributions and turn them into log.
numFeat = data.dtypes[data.dtypes != "object"].index
skewedFeat = data[numFeat].apply(lambda x: stats.skew(x))
skewedFeat=skewedFeat[skewedFeat>0.75].index
data[skewedFeat] = np.log1p(data[skewedFeat])
```

## Now to fit

```python
train_data = pd.get_dummies(data.loc[:,data.columns!='SalePrice'])
train_y = data['SalePrice']
```

```python
train_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, train_y)
```

```python
rdmFrst = RandomForestClassifier(n_estimators=100, random_state=1)
```

```python
rdmFrst.fit(X_train,y_train)
rdmFrst.score(X_test,y_test)
```

```python
def rmse_cv(model):
    rmse=np.sqrt(-cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
model_lasso = LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(X_train, y_train)
```

```python
rmse_cv(model_lasso)
```

```python
coef = pd.Series(model_lasso.coef_,index=X_train.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) 
      + " variables and eliminated the other " 
      +  str(sum(coef == 0)) + " variables")
```

```python
vis_coef = pd.concat([coef.nlargest(10),coef.nsmallest(10)])
```

```python
sns.barplot(x=vis_coef,y=vis_coef.index)
```

```python
sns.residplot(model_lasso.predict(X_train),y_train)
```

```python
model_lasso.score(X_test,y_test)
```
